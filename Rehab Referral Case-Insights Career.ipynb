```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
df=pd.read_excel('Rehab Referral Case-Insights Career.xlsx',sheet_name='data')
print('data has been successfully imported')
```

```python
df.columns
```

```python
#1 count the number of separations, sum up the total benefits paid, average charge price,count the number of
#unique hospitals by each hospital group,sorted by the descending order of average charging price$

df.groupby(['hospgrp'],as_index=False).agg({'clmseprtid':'count','totfundhospmedbenefitamt':'sum',\
                                            'episode28d_chg':'mean','prvnbr':'nunique'})\
.sort_values(by=['episode28d_chg'],ascending=[False])
```

```python
#2 count the number of separations, the average length of stay, average ICU days, average benefits paid by hospital 
#acquired complexity flag,sorted by the descending order of average benefit paid$

df.groupby(['hacofadj1_flg'],as_index=False).agg({'clmseprtid':'count','clmlosdays':'mean','clmicudays':'mean',\
                                                  'totfundhospmedbenefitamt':'mean'})\
.sort_values(by=['totfundhospmedbenefitamt'],ascending=[False])
```

```python
#3 count the number of patients did surgery by genders, and also count the number of patients did rehab by genders. 
#in the end, find rehab rate of each gender (rehab rate is the percentage % of patients choose rehab treatments 
#after their acute surgeries. )

table1=df.groupby(['patientgender']).agg({'clmseprtid':'count','rehab_clmseprtid':'count'})
table1['rehab_rate']=table1.rehab_clmseprtid/table1.clmseprtid
table1

```

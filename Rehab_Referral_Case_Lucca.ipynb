```python
df.columns
```

```python
# generate new table with required colomns
table1=df.groupby(['rehab_hospgrp'],as_index=False).agg({'rehab_clmseprtid':'count'}).\
sort_values(by=['rehab_clmseprtid'],ascending=False)
table1['hospgrpID']='HG'+table1.rehab_hospgrp.str[-2:]
table1
```

```python
#plot bar chart
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(20,5))
plt.bar(x=table1.hospgrpID,height=table1.rehab_clmseprtid,color='cornflowerblue')
plt.title('The number of rehab cases by each hospital group')
plt.xlabel('Hospital group')
plt.ylabel('The number of rehab cases over 3 years')

#add annotation
plt.annotate('Most popular hospital group',xy=(0,3900),xytext=(0,4500),\
             arrowprops=dict(facecolor='cornflowerblue',width=2),color='steelblue')
plt.show()
```

### Conclustion:
* Hospital group 39 and 9 are the most popular ones, with patients number more than 3000. The patients number of the hospital group in the 3rd place is significantly lower, with only 1387 patients.

### Objective:
* geograpic location by hospital group (highest AU coverage)

```python
df.groupby(['rehab_hospgrp'],as_index=False).agg({'hospstate':'unique','prvnbr':'nunique'}).\
sort_values(by=['prvnbr'],ascending=False)
```

### Conclusion:
* Hospital group 9 and 39 also have the widest coverage in terms of the number of hospitals and state coverage. Since popularity and coverage are important metrics, which determine the quality of our cooperation, we could narrow our selection down to these two groups.

### Objective:
* rehab result by hospital group (highest FIM score change, assumption )

```python
#generate new table, add new columns
table1=df.groupby(['rehab_hospgrp'],as_index=False).agg({'totdischargefim':'mean',
       'totadmitfim':'mean'})
table1['fimdiff']=table1.totdischargefim-table1.totadmitfim
table1['hospID']='HP'+table1.rehab_hospgrp.str[-2:]
table2=table1.sort_values(by=['fimdiff'],ascending=False)
table2
```

```python
#plot bar chart
fig=plt.figure(figsize=(20,5))
plt.bar(x=table2.hospID,height=table2.fimdiff,color='darkkhaki')
plt.title('The FIM difference by hospital group')
plt.xlabel('Hospital group')
plt.ylabel('The FIM difference across 3 years')
plt.axhline(table2.fimdiff.mean(),ls='--',color='grey')

#add annotation
plt.annotate('Highest FIM difference',xy=(0,22.3),xytext=(0,26),arrowprops=dict(facecolor='darkkhaki',width=2),\
            color='darkgoldenrod')
plt.show()
```

### Conclusion:
* Hospital group 39 has higher FIM than hospital group 9, and its score is above average. The FIM score of hospital group 9 is below average, which is concerning because rehab quality is very important.

### Objective:
* average rehab benefits paid by hospital group 

```python
#generate new table and add new column
table1=df.groupby(['rehab_hospgrp'],as_index=False).agg({'rehab_HospExProsBen':'mean'}).\
sort_values(by=['rehab_HospExProsBen'],ascending=True)
table1['hospID']=table1.rehab_hospgrp.str[-2:]
table1
```

```python
#plot bar chart
fig=plt.figure(figsize=(20,5))
plt.bar(x=table1.hospID,height=table1.rehab_HospExProsBen,color='rosybrown')
plt.title('Average benefits paid by hospital group')
plt.xlabel('Hospital group')
plt.ylabel('Average benefits across 3 years')
plt.axhline(table1.rehab_HospExProsBen.mean(),ls='--',color='grey')
#add annotation
plt.annotate('Most financially efficiendt',xy=(0,3800),xytext=(0,5900),\
             arrowprops=dict(facecolor='rosybrown',width=2),color='indianred')
plt.show()
```

### Conclusion:
* The average benefits paid by hospital group 9 is higher than hospital group 39, which are all above average. Based on all the metrics mentioned above, we may conclude that hospital group 39 is our best choice. We'll look at some other metrics to justify our choice.

### Objective:
* average rehab length of stay by hospital group 

```python
#generate new table and add column
table1=df.groupby(['rehab_hospgrp'],as_index=False).agg({'rehab_ClmLOSDays':'mean'}).\
sort_values(by=['rehab_ClmLOSDays'],ascending=True)
table1['hospID']='HP'+table1.rehab_hospgrp.str[-2:]
table1
```

### Conclusion:
* The average rehab length of stay of hospital group 39 is slightly higher than the average, which is acceptable. We could conclude that hospital group 39 is our best cooperation partner. 

## Create a function to plot rehab rate

#### analysis on the factors on rehab rete--> find out what factors have impact on whether patient choose to do inpatient rehab.

* 1. gender
* 2. location
* 3. age group 
* 4. complexity
* 5. DRG

```python
import matplotlib.pyplot as plt
```

```python
def rehab_rate_plot(group):
    #find the rehab rate for each group
    table1=df.groupby([group],as_index=False).agg({'rehab_clmseprtid':'count','clmseprtid':'count'})
    table1['rehab_rate']=table1.rehab_clmseprtid/table1.clmseprtid
    table1=table1.sort_values(by=['rehab_rate'],ascending=False)
    print(table1)
    
    #plot bar chart
    plt.bar(x=table1[group],height=table1.rehab_rate)
    plt.show()
```

```python
rehab_rate_plot('patientgender')
```
